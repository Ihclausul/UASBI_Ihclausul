```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
caesarian_df = pd.read_excel("E:/1.Saya/Kuliah/Semester 7/BI/UAS/data_UAS-main/caesarian.xlsx",
                       usecols=['Usia','Kelahiran_ke-','Waktu_Kelahiran','Tekanan_darah','Kelainan_jantung', 'Caesarian'],
                             dtype={'Usia': 'int32', 'Kelahiran_ke-': 'int32', 'Waktu_Kelahiran': 'int32', 'Tekanan_darah': 'int32', 
                                    'Kelainan_jantung': 'int32', 'Caesarian': 'int32'})
```

```python
caesarian_df.head()
```

```python
import math

dis = []
for i in range(80):
    dis.append(math.sqrt((float(caesarian_df.iloc[i]['Usia'])-25)**2+
                         (float(caesarian_df.iloc[i]['Kelahiran_ke-'])- 1)**2+
                         (float(caesarian_df.iloc[i]['Waktu_Kelahiran'])-0)**2+
                         (float(caesarian_df.iloc[i]['Tekanan_darah'])-1)**2))
```

```python
caesarian_df['dis'] = dis
caesarian_df
```

```python
caesarian_df.sort_values('dis')
```

```python
y = caesarian_df.sort_values('dis').head(5)
y
```

```python
z = y["Caesarian"]
z
```

```python
np.mean(z)
```
