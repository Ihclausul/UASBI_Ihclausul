```python
import math

dis = []
for i in range(5000):
    dis.append(math.sqrt((float(df.iloc[i]['ID'])-4)**2+
                         (float(df.iloc[i]['Income'])- 49)**2+
                         (float(df.iloc[i]['Education'])-1)**2+
                         (float(df.iloc[i]['CreditCard'])-1)**2))
```

```python
df['dis'] = dis
df
```
